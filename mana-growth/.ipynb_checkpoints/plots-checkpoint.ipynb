# generate plots

Blocks for different plots

```python
import matplotlib.pyplot as plt
```

```python
tables = {}
d = 3
```

```python
for n in [3, 4]:
    key = 'h_' + str(n) + '_' + str(d)
    f = open('output/random_pure_' + str(n) + '_' + str(d), 'r')
    tables[key] = []
    while True:
        a = f.readline()
        if a == '':
            break
        tables[key].append(float(a))
    f.close()
    for t in [10 + 10*i for i in range(20)]:
        key = '1d_' + str(n) + '_' + str(d) + '_' + str(t)
        f = open('output/random_1d_' + str(n) + '_' + str(d) + '_' + str(t), 'r')
        tables[key] = []
        while True:
            a = f.readline()
            if a == '':
                break
            tables[key].append(float(a))
        f.close()
```

```python
plt.clf()
n, bins, patches = plt.hist(tables['1d_3_3_50'], 50, density=True, facecolor='b', alpha=0.75)
#plt.xlabel('Mana')
#plt.ylabel('Probability')
#plt.title('Mana distribution')
```
