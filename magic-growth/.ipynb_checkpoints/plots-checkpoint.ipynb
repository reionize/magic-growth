# generate plots

Blocks for different plots

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager
```

```python
tables = {}
d = 3
```

```python
for n in [3, 4, 5]:
    key = 'h_' + str(n) + '_' + str(d)
    f = open('output/random_pure_' + str(n) + '_' + str(d), 'r')
    tables[key] = []
    while True:
        a = f.readline()
        if a == '':
            break
        tables[key].append(float(a))
    f.close()
    for t in [10 + 10*i for i in range(20)]:
        key = '1d_' + str(n) + '_' + str(d) + '_' + str(t)
        f = open('output/random_1d_' + str(n) + '_' + str(d) + '_' + str(t), 'r')
        tables[key] = []
        while True:
            a = f.readline()
            if a == '':
                break
            tables[key].append(float(a))
        f.close()
```

```python
plt.clf()
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
n, bins, patches = plt.hist(tables['1d_3_3_60'], [0.017*i for i in range(96)], density=True, histtype='step', color='b', alpha=0.5, linewidth=1.5, label='t = 60')
n, bins, patches = plt.hist(tables['h_3_3'], bins, histtype='step', density=True, color='r', alpha=0.5, linewidth=1.5, label='Haar-random')
for tick in ax.get_xticklabels():
    tick.set_fontname("cmss10")
    #tick.set_fontname("cmr10")
#plt.xlabel('Mana of output state')
#plt.ylabel('Frequency')
#plt.title('Mana distribution')
#font = {'size':20, 'family':'serif', 'serif': ['computer modern roman']}
#plt.rc('font',**font)
fig.set_size_inches(6, 4)
fig.set_dpi(100)
frame = plt.gca()
frame.axes.get_yaxis().set_visible(False)
plt.rcParams.update({'font.size': 16})
plt.legend(bbox_to_anchor=(1.05, 0.75), loc='upper left')
#plt.grid(True)
#plt.show()
#plt.savefig('plots/comp_1d_3_3_60.pdf', bbox_inches='tight', pad_inches=0.5)
```

```python
from matplotlib import cm
cmap = cm.get_cmap('inferno')
```

```python
plt.clf()
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
bins = [0.017*i for i in range(96)]
for t in [10 + 50*i for i in range(4)]:
    plt.hist(tables['1d_3_3_'+str(t)], bins, density=True, histtype='step', color=cmap(1-t/200-20/200), alpha=0.75, linewidth=1.5, linestyle=(0, (5, 1)), label='t = '+str(t))
#plt.hist(tables['h_3_3'], bins, density=True, facecolor='r', alpha=0.5)
plt.hist(tables['h_3_3'], bins, histtype='step', density=True, color=cmap(0), alpha=0.75, linewidth=1.5, label='Haar-random')
for tick in ax.get_xticklabels():
    tick.set_fontname("cmss10")
    #tick.set_fontname("cmr10")
#plt.xlabel('Mana of output state')
#plt.ylabel('Frequency')
#plt.title('Mana distribution')
fig.set_size_inches(6, 4)
fig.set_dpi(100)
frame = plt.gca()
frame.axes.get_yaxis().set_visible(False)
plt.rcParams.update({'font.size': 16})
plt.legend(bbox_to_anchor=(1.05, 0.75), loc='upper left')
#plt.show()
plt.savefig('plots/comp_1d_3_3_60.pdf', bbox_inches='tight', pad_inches=1)
```

```python
import numpy as np
```

```python
for n in [3, 4, 5]:
    tables['avg_1d_'+str(n)+'_3'] = [np.average(tables['1d_'+str(n)+'_3_'+str(t)]) for t in [10 + 10*i for i in range(20)]]
    tables['avg_h_'+str(n)+'_3'] = np.average(tables['h_'+str(n)+'_3'])
    tables['std_1d_'+str(n)+'_3'] = [np.std(tables['1d_'+str(n)+'_3_'+str(t)], ddof=1) for t in [10 + 10*i for i in range(20)]]
    tables['std_h_'+str(n)+'_3'] = np.std(tables['h_'+str(n)+'_3'], ddof=1)
```

```python
plt.clf()
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
for n in [3, 4, 5]:
    plt.plot([10 + 10*i for i in range(20)], tables['avg_1d_'+str(n)+'_3'], color=cmap((n-3)/3), alpha=0.75, linewidth=1)
plt.xlabel('Circuit depth')
plt.ylabel('Average mana')
#plt.title('Title')
fig.set_size_inches(6, 4)
fig.set_dpi(100)
frame = plt.gca()
#frame.axes.get_yaxis().set_visible(False)
plt.rcParams.update({'font.size': 16})
plt.rcParams.update({'font.family': 'cmss10'})
plt.grid(True)
#plt.show()
#plt.savefig('plots/conv_1d_3_3.pdf')
```
